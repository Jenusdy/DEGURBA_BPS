```python
import geopandas as gpd
from glob import glob
```

```python
ls_=glob("data/output/1_population_density/grid_pop_*_1km_w_bu.gpkg")
```

```python
import pandas as pd
```

```python
data_=pd.DataFrame(ls_).rename(columns={0:'path'})
data_['idkab']=data_.path.apply(lambda y:y.split('/')[3].replace('grid_pop_','').replace('_1km_w_bu.gpkg',''))
```

```python
data_['prov']=data_['idkab'].apply(lambda y: y[:2])
```

```python
data_=data_.sort_values(['prov','idkab'])
```

```python
def sekm_ok(i):
    global data_
    t=data_.query('prov==@i')
    temp=gpd.GeoDataFrame()
    for j in t.path:
        temp1=gpd.read_file(j)[['p_builtup','pop','geometry']]
        temp=temp1.append(temp)
    temp['cenx']=temp.geometry.centroid.x.astype(str)+'_'+temp.geometry.centroid.y.astype(str)
    temp2=temp.dissolve(by='cenx',aggfunc='sum')
    temp2.to_file("data/output/1_5_population_density/1km_prov"+i+"_w_bu.gpkg")
    print(temp2)
```

```python
data_.query('prov.str.slice(0,1) not in ["1","2","3","5","6"]')
```

```python
from tqdm.contrib.concurrent import process_map
```

```python
process_map(sekm_ok,data_.prov.unique()) 
```

```python
data_['path_prov']=data_.prov.apply(lambda y: "data/output/1_5_population_density/1km_prov"+y+"_w_bu.gpkg")
```

```python
data_pulau=data_[['pulau','path_prov','prov']].drop_duplicates()
```

```python
def sepulau_ok(i):
    global data_pulau
    t=data_.query('pulau==@i')
    temp=gpd.GeoDataFrame()
    for j in t.path:
        temp1=gpd.read_file(j)[['p_builtup','pop','geometry']]
        temp=temp1.append(temp)
    temp['cenx']=temp.geometry.centroid.x.astype(str)+'_'+temp.geometry.centroid.y.astype(str)
    temp2=temp.dissolve(by='cenx',aggfunc='sum')
    temp2.to_file("data/output/1_5_population_density/00_pulau_"+i+"_w_bu.gpkg")
    print(temp2)
```

```python
process_map(sepulau_ok,data_pulau.pulau.unique()) 
```

```python
data_nas=gpd.read_file('data/input/nas1.gpkg')
```

```python
def finalize(dt):
    global data_nas
    print(dt)
    y_=gpd.read_file(dt)
    y_['geometry']=y_.geometry.centroid
    tes=gpd.sjoin(data_nas,y_,how='inner',predicate='intersects')[['geometry','p_builtup','pop']]
    tes['cenx']=tes.geometry.centroid.x.astype(str)+'_'+tes.geometry.centroid.y.astype(str)
    tes=tes.dissolve(by='cenx',aggfunc='sum')
    name=dt.replace('w_bu','finalized').replace('00_','00_fin_')
    tes.to_file(name)
```

```python
process_map(finalize,dt_gl) 
```
